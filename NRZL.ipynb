```python
import matplotlib.pyplot as plt

def nrzl_encode(data):
    encoded = []
    for bit in data:
        if bit == '1':
            encoded.append(1)
        else:
            encoded.append(0)
    return encoded

def plot_signal(bitstream):
    encoded = nrzl_encode(bitstream)
    time = []
    signal = []
    t = 0
    for bit in encoded:
        time += [t, t+1]
        signal += [bit, bit]
        t += 1

    plt.step(time, signal, where='post')
    plt.ylim(-0.5, 1.5)
    plt.text(0, 1.1, f"Input: {bitstream}")
    plt.grid(True)
    plt.show()

plot_signal('10111001')
```

```python
original_data = '10111001'

def nrzl_encode(data):
    encoded = []
    for bit in data:
        if bit == '1':
            encoded.append(1)
        else:
            encoded.append(0)
    return encoded

def nrzl_decode(encoded):
    decoded = []
    for level in encoded:
        if level == 1:
            decoded.append('1')
        else:
            decoded.append('0')
    return ''.join(decoded)

encoded_signal = nrzl_encode(original_data)
decoded_data = nrzl_decode(encoded_signal)

print("Original Data     :", original_data)
print("NRZ-L Encoded     :", encoded_signal)
print("NRZ-L Decoded     :", decoded_data)
```
