```python
import matplotlib.pyplot as plt

def mlt3_encode(data):
    b=[0,1,0,-1]
    idx=0
    encoded=[]
    pulse=0
    for bit in data:
        if bit=='1':
            idx=idx+1
            pulse=b[idx%4]

        encoded.append(pulse)

    return encoded

def plot_input_signal(bitstream):
    time = []
    signal = []
    t = 0

    for bit in bitstream:
        val = 1 if bit == '1' else 0
        time += [t, t + 1]
        signal += [val, val]
        t += 1

    plt.figure(figsize=(10, 2.5))
    plt.xticks(range(len(bitstream) + 1))
    plt.step(time, signal, where='post', color='gray')
    
    for i, bit in enumerate(bitstream):
        plt.text(i + 0.5, 1.1, bit, ha='center')
        
    plt.ylim(-0.5, 1.5)
    plt.grid(True)
    plt.show()


def plot_signal(bitstream):
    encoded=mlt3_encode(bitstream)
    time=[]
    signal=[]
    t=0
    for bit in encoded:
        time+=[t,t+1]
        signal+=[bit,bit]
        t+=1
        
    plt.figure(figsize=(10, 2.5))
    plt.step(time, signal,where='post')
    plt.ylim(-1.5,1.5)
    plt.xticks(range(len(bitstream) + 1))
    plt.grid(True)
    plt.show()

str=input("Enter the Bitstream : ")
plot_input_signal(str)
plot_signal(str)

def mlt3_decode(encoded):
    decoded = ''
    prev = 0
    for level in encoded:
        if level == prev:
            decoded += '0'
        else:
            decoded += '1'
        prev = level

    return decoded

encoded = mlt3_encode(str)
print("Encoded:", encoded)
decoded = mlt3_decode(encoded)
print("Decoded:", decoded)

```
