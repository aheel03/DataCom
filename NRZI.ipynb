```python
import matplotlib.pyplot as plt

def nrzi_encode(data):
    encoded=[]
    level=1
    for bit in data:
        if bit=='1':
            level^=1
        encoded.append(level)
    return encoded

def plot_signal(bitstream):
    encoded=nrzi_encode(bitstream)
    time=[]
    signal=[]
    t=0
    for bit in encoded:
        time+=[t,t+1]
        signal+=[bit,bit]
        t+=1

    plt.step(time, signal,where='post')
    plt.ylim(-0.5,1.5)
    plt.text(0,1.1,f"Input: {bitstream}")
    plt.grid(True)

    plt.show()

plot_signal('10111001')
```

```python
original_data = '10111001'

def nrzi_encode(data):
    encoded = []
    level = 1
    for bit in data:
        if bit == '1':
            level ^= 1
        encoded.append(level)
    return encoded

def nrzi_decode(encoded):
    decoded = []
    prev = encoded[0]
    decoded.append('1')
    for current in encoded[1:]:
        if current != prev:
            decoded.append('1')
        else:
            decoded.append('0')
        prev = current
    return ''.join(decoded)

encoded_signal = nrzi_encode(original_data)
decoded_data = nrzi_decode(encoded_signal)

print("Original Data     :", original_data)
print("NRZ-I Encoded     :", encoded_signal)
print("NRZ-I Decoded     :", decoded_data)
```
