```python
import matplotlib.pyplot as plt

def ami_encode(data):
    encoded=[]
    level=1
    for bit in data:
        if bit=='0':
            if level==1:
                level=-1
            else:
                level=1 
            encoded.append(level)
        else:
            encoded.append(0)
            
    return encoded

def plot_input_signal(bitstream):
    time = []
    signal = []
    t = 0

    for bit in bitstream:
        val = 1 if bit == '1' else 0
        time += [t, t + 1]
        signal += [val, val]
        t += 1

    plt.figure(figsize=(10, 2.5))
    plt.xticks(range(len(bitstream) + 1))
    plt.step(time, signal, where='post', color='gray')
    
    for i, bit in enumerate(bitstream):
        plt.text(i + 0.5, 1.1, bit, ha='center')
        
    plt.ylim(-0.5, 1.5)
    plt.grid(True)
    plt.show()


def plot_signal(bitstream):
    encoded=ami_encode(bitstream)
    time=[]
    signal=[]
    t=0
    for bit in encoded:
        time+=[t,t+1]
        signal+=[bit,bit]
        t+=1
    
    plt.figure(figsize=(10, 2.5))
    plt.xticks(range(len(bitstream) + 1))
    plt.step(time, signal,where='post')
    plt.ylim(-1.5,1.5)
    plt.text(0,1.1,f"Input: {bitstream}")
    plt.grid(True)

    plt.show()

str=input("Enter the Bitstream : ")
plot_input_signal(str)
plot_signal(str)

def ami_decode(encoded):
    decoded = ''
    for voltage in encoded:
        if voltage == 0:
            decoded += '1'
        else:
            decoded += '0'
    return decoded

encoded = ami_encode(str)
print("Encoded:", encoded)
decoded = ami_decode(encoded)
print("Decoded:", decoded)

```
